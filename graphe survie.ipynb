```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4] , [5,6,7,8] , "b-" , linewidth=0.8 , marker="*" , label="111111111")
plt.plot([1,2,3,4] , [8,7,6,5] , "ks-" , linewidth=0.5)
plt.xlabel('bleu')
plt.ylabel('orange')
plt.title('couleur')
plt.axis([0,5,4,9])
plt.text(2,5.7,'boom')
plt.show()
```

graphe montrant le nombre de jours survécus en fonction de la capadap initiale avec une force initiale de 0.5

```python
name = ['-18', '18-25', '25-50', '50+']
data = [5000, 26000, 21400, 12000]

explode=(0, 0.15, 0, 0)
plt.pie(data, explode=explode, labels=name, autopct='%1.1f%%', startangle=90, shadow=True)
plt.axis('equal')
plt.show()

```

```python
name = ['entre 0 et 2', 'entre 2 et 4', 'entre 4 et 6', 'entre 6 et 8','entre 8 et 10']
data = [5000, 26000, 21400, 12000, 8000]

explode=(0.1, 0.15, 0.1, 0.1, 0.1)
plt.pie(data, explode=explode, labels=name, autopct='%1.1f%%', startangle=0, shadow=True)
plt.axis('equal')
plt.title("jour en fonction du biome")
plt.show()
```
